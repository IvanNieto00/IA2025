```python
# Generamos las funciones de pertenencia difusas
lo = fuzz.gbellmf(x, 6, 4, -10)
md = fuzz.gbellmf(x, 4, 4, 0)
hi = fuzz.gbellmf(x, 6, 4, 10)
```

```python
# Graficamos la calidad de la comida
fig, ax0 = plt.subplots(figsize=(8, 3))

ax0.plot(x, lo, 'b', linewidth=1.5, label='Pequeño')
ax0.plot(x, md, 'g', linewidth=1.5, label='Mediano')
ax0.plot(x, hi, 'r', linewidth=1.5, label='Grande')
ax0.set_title('Sugeno 2° orden')
ax0.legend()

plt.tight_layout()
```

```python
lista = []
for i in range(-10,10,1):
    lolo = fuzz.interp_membership(x, lo, i0)
    lololo = i*i+0.1 * i + 6.4
    mdmd = fuzz.interp_membership(x, md, i)
    mdmdmd = i*i-0.5 * i + 4
    hihi = fuzz.interp_membership(x, hi, i)
    hihihi = 1.8 * i*i+ i -2
    print(i, lololo*lolo + mdmdmd*mdmd + hihihi*hihi)
    lista.append(lololo*lolo + mdmdmd*mdmd + hihihi*hihi)
```

```python
plt.figure()
plt.plot(range(-10, 10, 1), lista)
plt.xlabel('Entrada (x)')
plt.ylabel('Salida (Sugeno)')
plt.title('Salida Modelo Sugeno')
plt.grid(True)
plt.show()
```
